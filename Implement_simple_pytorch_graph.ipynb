## We will try to implement the above pytorch model using simple classes with the help of pytorch methods

```python
import torch
from torchviz import make_dot
```

```python
class SimpleGraph:
    
    def __init__(self,w,b):
        '''
        Initializing a Simple graph with weight and bias
        
        Args:
            w(float) : initial value for weights
            b(float) : initial value for bias
        '''
        
        self.w = torch.tensor([w],requires_grad = True)
        self.b = torch.tensor([b],requires_grad = True)
        
        
    def forward(self,X):
        
        '''
        Make a forward pass with input features
        
        Args:
            X = 1D tensor of features
        '''
        predictions = torch.tanh(self.w * X + self.b)
        
        return predictions
    
def sq_loss(y_hat,y):
        '''
        Calculate square loss fn
        
        Args
            y_hat = predictions
            y = true values
            both 1d tensor
        '''
        
        loss = torch.square(y-y_hat) * 0.5
        
        return loss
        
        
    
        
        
```

```python

feature = torch.tensor([1.0])  # input tensor
target = torch.tensor([7.0])  # target tensor

simple_graph = SimpleGraph(-0.5, 0.5)
print(f"initial weight = {simple_graph.w.item()}, "
      f"\ninitial bias = {simple_graph.b.item()}")

prediction = simple_graph.forward(feature)
square_loss = sq_loss(target, prediction)

print(f"for x={feature.item()} and y={target.item()}, "
      f"prediction={prediction.item()}, and L2 Loss = {square_loss.item()}")

make_dot(square_loss)
```
